```python
class VacuumCleanerAgent:
    def __init__(self):
        self.environment = {"A": 0, "B": 0}  # assumed initial state

    def set_environment(self):
        print("Enter dirt status for location A (0 for clean, 1 for dirty): ")
        self.environment["A"] = int(input())
        print("Enter dirt status for location B (0 for clean, 1 for dirty): ")
        self.environment["B"] = int(input())
        print("\nInitial environment: ", self.environment)

    def clean(self, location):
        if self.environment[location] == 1:
            print(f"Location {location} is dirty.")
            print(f"Vacuum cleaner cleaned the dirt at {location}.")
            self.environment[location] = 0
        else:
            print(f"Location {location} is already clean.")

    def clean_at_location(self, location):
        print(f"\nVacuum cleaner at {location} location.")
        self.clean(location)
        other_location = "B" if location == "A" else "A"
        if self.environment[other_location] == 1:
            print(f"Moving to {other_location} location.")
            self.clean(other_location)

    def start_cleaning(self):
        if self.environment["A"] == 0 and self.environment["B"] == 0:
            print("\nBoth the locations are clean.")
        else:
            if self.environment["A"]:
                self.clean_at_location("A")
            else:
                self.clean_at_location("B")


if __name__ == "__main__":
    vacuum_agent = VacuumCleanerAgent()
    vacuum_agent.set_environment()  # Take user input for initial environment
    vacuum_agent.start_cleaning()   # Start the cleaning process
```
